```python
q=[]
def do_list(a):
    q.append(a)
    print(f"список: {q}")
print('Вы вводите элементы в множество "a". Введите "exit" чтобы закончить набор')
while True:
    w=input('Введите число, которое хотели бы занести в список: ')  
    if w!='exit':
        do_list(w)
    else:
        break
a=set(q)
q=[]
print('Вы вводите элементы в множество "b". Введите "exit" чтобы закончить набор')
while True:
    w=input('Введите число, которое хотеи бы занести в список: ')  
    if w!='exit':
        do_list(w)
    else:
        break
b=set(q)

c=a&b
r=list(c)
r.sort()
print(f"Элементы,встречающиеся в двух списках, по возрастанию: {r}")
```
